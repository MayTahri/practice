
### Recap: From equations to lines

Any line is given by the equation: 

$$
w_0 + \vec{w}\cdot{\vec{x}} = 0.
$$ 


Consider the case of $D=2$, where $\vec{w}=\begin{bmatrix}w_1\\w_2\end{bmatrix}$, and $\vec{x}=\begin{bmatrix}x_1\\x_2\end{bmatrix}$. 

This is equivalent to:

$$
w_0 + w_1x_1 + w_2x_2 = 0 
$$

**Plotting the line**

If we rearrange, we see how e.g. $x_2$ changes when $x_1$ is changed: 

$$x_2 = \frac{-w_0 -w_1x_1}{w_2}$$

We can use this as the basis of plotting the line: generate values for $x_1$ and insert them into the formula
to obtain the corresponding $x_2$ that satisfy the linear relation. A simple example is below. 

```python
import matplotlib.pyplot as plt
import numpy as np


def line(x, w_0, w):
    return -(w_0 + w[0] * x) / w[1]

# Write a function to plot a line for $D=2$, given $w_0$ and $\vec{w}$, and the starting and ending $x_1$ values. 
# (see linear_utils.py for a more advanced version that ensures the plot is square)

def plot_line(w_0, w, x1_start, x1_end):
    x = np.array([x1_start, x1_end])
    y = line(x, w_0, w)

    plt.plot(x, y)

# Example: w_0=-1, w=[2, 3], x1_start=-2, x1_end=2
    
plot_line(-1, [3, -2], -2, 2)
```

A nice property of the equation of the line is that $\vec{w}$ gives the **direction** of the line, and $w_0$ its position:

```python
# A nice property of the equation of the line is that $\vec{w}$ gives the **direction** of the line, and $w_0$ its position:
import linear_utils as lu

plt.figure(figsize=(5, 5))

w_0 = 0
w = [1, 2]

# Plot line (with grid)
lu.plot_line_square(w_0, w, -3, 3)

# Plot w vector
plt.arrow(0, 0, w[0], w[1], head_width=0.2, head_length=0.2, fc='black', ec='black')
```

- $\vec{w}$ is a vector pointing orthogonally to the line in the direction where $w_0 + \vec{w}\cdot\vec{x}>0$.
- $w_0$ shifts the line (see Figure below)

```python
plt.figure(figsize=(5, 5))

w_0 = 3
w = [1, 2]

# Plot line (with grid)
lu.plot_line_square(0, w, -3, 3, style="g--", label='$w_0=0$')
lu.plot_line_square(w_0, w, -3, 3, label='$w_0=3$')

# Plot w vector on top of one of the points
x1 = 0
x2 = line(0, w_0, w)
plt.arrow(x1, x2, w[0], w[1], head_width=0.2, head_length=0.2, fc='black', ec='black')

plt.legend()
```
